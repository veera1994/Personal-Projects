## Handling missing values

```python
df.columns
```

```python
df = df.drop(['Ticket', 'Name', 'PassengerId', 'Cabin'], axis=1)
```

```python
df.Sex.value_counts()
```

```python
df['Sex'] = df.Sex.map({'male':0, 'female':1})
```

```python
df.isnull().sum()
```

```python
df.Age = df.groupby(['Pclass','Sex'])['Age'].transform(lambda x:x.fillna(x.mean()))
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

## Feature engineering

```python
df.describe()
```

```python
bin_labels = [1,2,3,4,5]
df['Age_group'] = pd.cut(x=df.Age, bins=5,labels=bin_labels)
```

```python
sns.countplot(data=df, x=df['Age_group'], hue='Survived')
```

```python
df.groupby('Age_group')['Survived'].mean()
```

```python
df = df.drop(['Age'], axis=1)
```

```python
df.Age_group = df.Age_group.astype(int)
```

```python
df['Family_size'] = df['SibSp'] + df['Parch'] + 1
```

```python
df.groupby('Family_size')['Survived'].mean()
```

```python
df['Is_alone'] = np.where(df.Family_size==1, 1, 0)
```

```python
df.groupby('Is_alone')['Survived'].mean()
```

```python
df = df.drop(['Parch', 'SibSp','Family_size'], axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
bin_labels = [1,2,3,4]
df['Fare'] = pd.qcut(df.Fare, 4, labels=bin_labels)
```

```python
df.groupby(['Fare'])['Survived'].mean()
```

```python
df['Fare'] = df['Fare'].astype(int)
```

```python
df.head()
```

```python
X = df.drop('Survived', axis=1)
y = df.Survived
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix, plot_roc_curve
import shap
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
decision_tree = DecisionTreeClassifier()
decision_tree.fit(X_train, y_train)
preds = decision_tree.predict(X_test)
```

```python
print(classification_report(y_test, preds))
```

```python
plot_confusion_matrix(decision_tree, X_test, y_test)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
preds = rf.predict(X_test)
print(classification_report(y_test, preds))
```

```python
plot_confusion_matrix(rf, X_test, y_test)
```

```python
feature_importance = rf.feature_importances_
features = pd.DataFrame(index=X_train.columns, data=feature_importance, columns=['importance'])
features = features.reset_index()
features.columns = ['Features', 'Importance']
features = features.sort_values(by='Importance', ascending=False)
sns.barplot(data=features, x='Features', y='Importance')
plt.xticks(rotation=90);
```

```python
explainer = shap.TreeExplainer(rf)
start_index = 3
end_index = 4
shap_values = explainer.shap_values(X_test[start_index:end_index])
X_test[start_index:end_index]
```

```python
shap.initjs()
preds = rf.predict(X_test[start_index:end_index])[0]
print(f'Predicted value is: {preds}')
shap.force_plot(explainer.expected_value[1], shap_values[1], X_test[start_index:end_index])
```

```python
shap_values = shap.TreeExplainer(rf).shap_values(X_test)
shap.summary_plot(shap_values, X_test)
```

```python
xg = XGBClassifier(learning_rate=0.01,n_estimators=3460,
                                     max_depth=3, min_child_weight=0,
                                     gamma=0, subsample=0.7,
                                     colsample_bytree=0.7,
                                     objective='reg:linear', nthread=-1,
                                     scale_pos_weight=1, seed=27,
                                     reg_alpha=0.00006)
xg.fit(X_train, y_train)
preds = xg.predict(X_test)
print(classification_report(y_test, preds))
```

```python
plot_confusion_matrix(xg, X_test, y_test)
```

```python
feature_importance = xg.feature_importances_
features = pd.DataFrame(index=X_train.columns, data=feature_importance, columns=['importance'])
features = features.reset_index()
features.columns = ['Features', 'Importance']
features = features.sort_values(by='Importance', ascending=False)
sns.barplot(data=features, x='Features', y='Importance')
plt.xticks(rotation=90);
```

```python
explainer = shap.TreeExplainer(xg)
start_index = 3
end_index = 4
shap_values = explainer.shap_values(X_test[start_index:end_index])
X_test[start_index:end_index]
```

```python
shap.initjs()
preds = xg.predict(X_test[start_index:end_index])[0]
print(f'Predicted value is: {preds}')
shap.force_plot(explainer.expected_value, shap_values, X_test[start_index:end_index])
```
