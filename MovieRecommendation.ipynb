## Demographic filtering

We are using IMDBs weighted rating to calculate a score for the movies

```python
# Mean vote across the dataset

C = df.vote_average.mean()
print(C)
```

```python
# Mininmum number of votes required to be in the list
m = df.vote_count.quantile(.9)
print(m)
```

```python
# Filtering the movies based on the minimum number of votes criteria
mr1 = df[df.vote_count >=m]
mr1.shape
```

```python
# Weighted metric
def index(x, m=m, C=C):
    v = x['vote_count']
    R = x['vote_average']
    return (((v/(v+m)) * R) + ((m/(m+v))*C))
```

```python
mr1['score'] = mr1.apply(index, axis=1)
```

```python
#Sort movies based on score calculated above
mr1 = mr1.sort_values('score', ascending=False)

#Print the top 15 movies
mr1[['title', 'vote_count', 'vote_average', 'score']].head(10)
```

```python
sns.barplot(data=mr1.sort_values(by='popularity', ascending=False).head(6), x='title', y='popularity')
plt.xticks(rotation=90);
```
